```python
import pandas as pd
from datetime import datetime, timedelta
import csv
```

```python
# Reading the dataset
zulo_bank = pd.read_csv(r'dataset\zulo_bank.csv')
```

```python
# Displaying the first 5 rows.
display(zulo_bank.head())
```

```python
# Check the data info

zulo_bank.info()
```

```python
# Fill up missing values with appropriate parameter
zulo_bank.fillna({
    'loanAmount': 0.0,
    'loanType': 'Unknown',
    'InterestRate': 0.0
}, inplace=True)
```

```python
# Check the data info

zulo_bank.info()
```

```python
# Convert to 1NF
#Splitting the 'fullnam' into two parts 'first_name' and 'last_name'

zulo_bank[['first_name', 'last_name']] = zulo_bank['FullName'].str.split(expand=True)  

display(zulo_bank.head())
```

```python
# Convert 1NF to 2NF
# Customer table

customer = zulo_bank[['first_name', 'last_name', 'Email', 'Phone']].copy().drop_duplicates().reset_index(drop=True)
customer['customer_id'] = range(1, len(customer) + 1)
customer = customer[['customer_id', 'first_name', 'last_name', 'Email', 'Phone']]
```

```python
customer
```

```python
# Account Table

account = zulo_bank[['AccountType', 'Balance', 'OpeningDate']].copy()  
account['account_id'] = range(1, len(account) +1)
account = account[['account_id', 'AccountType', 'Balance', 'OpeningDate']]
```

```python
# Transaction Table

transaction = zulo_bank[['TransactionType', 'Amount', 'TransactionDate']].copy().drop_duplicates().reset_index(drop=True)
transaction['transaction_id'] = range(1, len(transaction) + 1)
transaction = transaction[['transaction_id', 'TransactionType', 'Amount', 'TransactionDate']] 
```

```python
loan = zulo_bank[['LoanAmount', 'LoanType', 'StartDate', 'EndDate', 'InterestRate']].copy().drop_duplicates().reset_index()
loan['loan_id'] = range(1, len(loan) + 1)
loan = loan[['loan_id', 'LoanAmount', 'LoanType', 'StartDate', 'EndDate', 'InterestRate']]
```

```python
loan
```

```python
# Merge opertaions to create the Zulo Bank

zulo_bank = zulo_bank.merge(customer, on=['first_name', 'last_name', 'Email', 'Phone'], how='left') \
                     .merge(account, on=['AccountType', 'Balance', 'OpeningDate'], how='left') \
                     .merge(transaction, on=['TransactionType', 'Amount', 'TransactionDate']) \
                     .merge(loan, on=['LoanAmount', 'LoanType', 'StartDate', 'EndDate', 'InterestRate']) \
                     [['customer_id', 'account_id', 'transaction_id', 'loan_id']]
```

```python
zulo_bank
```

```python
# Convert 2NF to 3NF
# Create the date dimension table
# Define the start and end dates

start_date = datetime(2020, 1, 1)
current_date = datetime(2020, 12, 31)

# Calculate the number of days between start date and current date
num_days = (current_date - start_date).days

# Generate a list of dates from start date to current date
date_list = [start_date + timedelta(days=x) for x in range(num_days + 1)]

# Ensure date_id matches the length of the date list
date = {'date_id': [x for x in range(1, len(date_list) + 1)], 'date': date_list}

# Create dataframe
date_dim = pd.DataFrame(date)
date_dim['Year'] = date_dim['date'].dt.year
date_dim['Month'] = date_dim['date'].dt.month
date_dim['Day'] = date_dim['date'].dt.day
date_dim['date'] = pd.to_datetime(date_dim['date']).dt.date
```

```python
date_dim
```

```python
# Account Table 2NF to 3NF
account['OpeningDate'] = pd.to_datetime(account['OpeningDate']).dt.date
account = account.merge(date_dim, left_on='OpeningDate', right_on='date', how='inner') \
                 .rename(columns={'date_id':'OpeningDate_ID'}) \
                 [['account_id', 'AccountType', 'Balance', 'OpeningDate_ID']]
```

```python
# Transaction Table 2NF to 3NF
transaction['TransactionDate'] = pd.to_datetime(transaction['TransactionDate']).dt.date
transaction = transaction.merge(date_dim, left_on='TransactionDate', right_on='date', how='inner') \
                         .rename(columns={'date_id': 'TransactionDate_ID'}) \
                         .reset_index(drop=True) \
                         [['transaction_id', 'TransactionType', 'Amount', 'TransactionDate_ID']]
```

```python
# Loan Table 2NF to 3NF

loan['StartDate'] = pd.to_datetime(loan['StartDate'], errors='coerce').dt.date
loan['EndDate'] = pd.to_datetime(loan['EndDate'], errors='coerce').dt.date

loan = loan.merge(date_dim, left_on='StartDate', right_on='date', how='inner') \
           .rename(columns={'date_id': 'StartDate_ID'}) \
           .merge(date_dim, left_on='EndDate', right_on='date', how='inner', suffixes=('', '_end')) \
           .rename(columns={'date_id': 'EndDate_ID'}) \
           [['loan_id', 'LoanAmount', 'LoanType', 'StartDate_ID', 'EndDate_ID', 'InterestRate']]
```

```python
# Save to our directory

transaction.to_csv(r'dataset\database_model\transactions.csv', index=False)
loan.to_csv(r'dataset\database_model\loan.csv', index=False)
account.to_csv(r'dataset\database_model\accounts.csv', index=False)
customer.to_csv(r'dataset\database_model\customers.csv', index=False)
date_dim.to_csv(r'dataset\database_model\date_dim.csv', index=False)
zulo_bank.to_csv(r'dataset\database_model\fact_table.csv', index=False)
```

```python
# Define database connection parameters including the database name
db_params = {
    'username': 'postgres',
    'password': 'MongoDB4luv',
    'host': 'localhost',
    'port': '5432',
    'database': 'Zulo_Bank'
}

default_db_url = f"postgresql://{db_params['username']}:{db_params['password']}@{db_params['host']}:{db_params['port']}/postgres"

# Attempt to create a database
try:  
    # Open the database
    conn = psycopg2.connect(default_db_url)
    conn.autocommit = True
    cur = conn.cursor()
    # Check if the database already exists
    cur.execute(f"SELECT 1 FROM pg_catalog.pg_database WHERE datname = '{db_params['database']}'")
    exists = cur.fetchone()
    if not exists:
        # Create the database
        cur.execute(f"CREATE DATABASE {db_params['database']}")
        print(f"Database {db_params['database']} created successfully.")
    else:
        print(f"Database {db_params['database']} already exists.")
    
    # Close the connection
    cur.close()
    conn.close()
except Exception as e:
    print(f"An error occured: {e}")
```

```python
# Connect to the database to create schema and tables.
def get_db_connection():
    connection = psycopg2.connect(
        host= 'localhost',
        database= 'zulo_bank',
        user= 'postgres',
        password= 'MongoDB4luv'
    )
    return connection

conn = get_db_connection()
```

```python
# Create schema and tables

def create_tables():
    conn = get_db_connection()
    cursor = conn.cursor()
    create_table_query = '''
                            CREATE SCHEMA IF NOT EXISTS zulo_bank;

                            DROP TABLE IF EXISTS zulo_bank.transactions CASCADE;
                            DROP TABLE IF EXISTS zulo_bank.accounts CASCADE;
                            DROP TABLE IF EXISTS zulo_bank.customers CASCADE;
                            DROP TABLE IF EXISTS zulo_bank.date_dim CASCADE;
                            DROP TABLE IF EXISTS zulo_bank.loans CASCADE;
                            DROP TABLE IF EXISTS zulo_bank.zulo_fact_table CASCADE;

                            CREATE TABLE IF NOT EXISTS zulo_bank.date_dim (
                                date_id SERIAL PRIMARY KEY,
                                date VARCHAR(10000),
                                year INT,
                                month INT,
                                day INT
                            );

                            CREATE TABLE IF NOT EXISTS zulo_bank.transactions (
                                transaction_id SERIAL PRIMARY KEY,
                                TransactionType VARCHAR(10000),
                                Amount FLOAT,
                                TransactionDate_ID INT,
                                FOREIGN KEY (TransactionDate_ID) REFERENCES zulo_bank.date_dim(date_id)
                            );

                            CREATE TABLE IF NOT EXISTS zulo_bank.accounts (
                                account_id SERIAL PRIMARY KEY,
                                AccountType VARCHAR(10000),
                                Balance FLOAT,
                                OpeningDate_ID INT,
                                FOREIGN KEY (OpeningDate_ID) REFERENCES zulo_bank.date_dim(date_id)
                            );

                            CREATE TABLE IF NOT EXISTS zulo_bank.loans (
                                loan_id SERIAL PRIMARY KEY,
                                LoanAmount FLOAT,
                                LoanType VARCHAR(10000),
                                StartDate_ID INT,
                                EndDate_ID INT,
                                InterestRate FLOAT,
                                FOREIGN KEY (StartDate_ID) REFERENCES zulo_bank.date_dim(date_id),
                                FOREIGN KEY (EndDate_ID) REFERENCES zulo_bank.date_dim(date_id)
                            );

                            CREATE TABLE IF NOT EXISTS zulo_bank.customers (
                                customer_id SERIAL PRIMARY KEY,
                                first_name VARCHAR(10000),
                                last_name VARCHAR(10000),
                                Email VARCHAR(10000),
                                Phone VARCHAR(10000)
                            );

                            CREATE TABLE IF NOT EXISTS zulo_bank.zulo_fact_table (
                                customer_id INT,
                                account_id INT,
                                transaction_id INT,
                                loan_ID INT,
                                FOREIGN KEY (customer_id) REFERENCES zulo_bank.customers(customer_id),
                                FOREIGN KEY (account_id) REFERENCES zulo_bank.accounts(account_id),
                                FOREIGN KEY (transaction_id) REFERENCES zulo_bank.transactions(transaction_id),
                                FOREIGN KEY (loan_id) REFERENCES zulo_bank.loans(loan_id)
                            );

                        '''
    
    cursor.execute(create_table_query)
    conn.commit()
    cursor.close()
    conn.close()
```

```python
create_tables()
```

```python
# Loading the data
# date_dim table

def load_data_from_csv(csv_path):
    conn = get_db_connection()
    cursor = conn.cursor()
    with open(csv_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            cursor.execute(
                '''INSERT INTO zulo_bank.date_dim (date_id, date, Year, Month, Day)
                   VALUES (%s, %s, %s, %s, %s)''',
                   row
            )
    
    conn.commit()
    cursor.close()
    conn.close()

# Provide the csv path to the file
csv_file_path = r'dataset\database_model\date_dim.csv'

load_data_from_csv(csv_file_path)
print('date_dim data loaded successfully')
```

```python
# Loading the data
# fact table

def load_data_from_csv(csv_path):
    conn = get_db_connection()
    cursor = conn.cursor()
    with open(csv_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            try:
                cursor.execute(
                '''INSERT INTO zulo_bank.zulo_fact_table (customer_id, account_id, transaction_id, loan_id)
                   VALUES (%s, %s, %s, %s, %s)''',
                   row
                )
            except psycopg2.IntegrityError:
                conn.rollback()
            else:
                conn.commit()
    cursor.close()
    conn.close()

# Provide the csv path to the file
csv_file_path = r'dataset\database_model\fact_table.csv'

load_data_from_csv(csv_file_path)
print('zulo_fact_table data loaded successfully')
```
